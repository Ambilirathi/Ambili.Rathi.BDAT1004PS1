```
//4th Question :Dynamic way (but its not taking input)
// Output of this code is uploaded

Array_Input();
 void Array_Input()
{
    Console.WriteLine("Enter the number of elements to be stored in the array: ");
    int elem = Convert.ToInt32(Console.ReadLine());
    //Initializing
    int[] Array_for_input = new int[elem]; 
    int[] Array_for_even = new int[elem]; 
    int[] Array_for_odd = new int[elem];
    int even_array_Index = 0;
    int odd_array_Index = 0;
    //Reading dynamic values to array 
    for (int i = 0; i < elem; i++)
    {
        Console.Write("Enter element - " + i + ": ");
        Array_for_input[i] = Convert.ToInt32(Console.ReadLine());
        Console.Write("\n");
    }
    //checking values is odd or even
    for (int i = 0; i < elem; i++)
    {
        if (Array_for_input[i] % 2 == 0)
        {
            Array_for_even[even_array_Index] = Array_for_input[i];
            even_array_Index++;
        }
        else
        {
            Array_for_odd[odd_array_Index] = Array_for_input[i];
            odd_array_Index++;
        }
    }
    //Displaying values as odd and even
    Console.WriteLine("The Even elements are: ");
    for (int i = 0; i < even_array_Index; i++)
    {
        Console.Write(Array_for_even[i] + " ");
    }
    Console.Write("\n");

    Console.WriteLine("The Odd elements are: ");
    for (int i = 0; i < odd_array_Index; i++)
    {
        Console.Write(Array_for_odd[i] + " ");
    }
    Console.Write("\n");
}
```
